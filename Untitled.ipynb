```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import cv2
import math
```

```python
ds, info = tfds.load('flic', split='train', with_info=True)
```

```python
example = ds.take(12)
print(example)
```

```python
debug = False
plt.figure(figsize=(12, 10))
for data in example:
    '''
        currframe
        image
        moviename
        poselet_hit_idx
        torsobox
        xcoords
        ycoords
    '''

    plt.clf()
    
    image = data['image'].numpy()
    shape = image.shape
    
    torsobox = data['torsobox'].numpy()
    pt1 = (int(torsobox[1]*shape[1]), int(torsobox[0]*shape[0]))
    pt2 = (int(torsobox[3]*shape[1]), int(torsobox[2]*shape[0]))

    cv2.circle(image, pt1, 5, (0, 255, 255), -1)
    cv2.circle(image, pt2, 5, (0, 255, 255), -1)
#     cv2.rectangle(image, torsobox, (0, 255, 255), 2, 1)
    
    xcoords = data['xcoords'].numpy()
    ycoords = data['ycoords'].numpy()
    xycoords = []
    for xy in zip(xcoords, ycoords):
        if not math.isnan(xy[0]):
            xycoords.append(xy)
    
    for x, y in xycoords:
        plt.scatter(x, y)
    plt.imshow(image)
    
    keys = ['currframe', 'moviename', 'poselet_hit_idx', 'torsobox', 'xcoords', 'ycoords']
    if debug == True:
        for key in keys:
            print(key + ':\n', data[key].numpy(), '\n')
```
